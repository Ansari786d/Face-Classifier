#  training and test set paths

```python
training_set = '../input/beauty-detection-data-set/train/'
test_set = '../input/beauty-detection-data-set/test/'
```

```python
import tensorflow as tf
```

```python
from  keras.preprocessing.image import ImageDataGenerator
```

# Applying Data Augmentation on Images 

```python
train_datagen = ImageDataGenerator(rescale=1/255,
                                  shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)
training_set= train_datagen.flow_from_directory('../input/beauty-detection-data-set/train/',
                                                target_size=(64,64),
                                                batch_size=16,
                                                class_mode = 'binary')

```

```python
test_gen = ImageDataGenerator(rescale=1./255)

test_set = test_gen.flow_from_directory('../input/beauty-detection-data-set/test/',
                                       target_size=(64,64),
                                       batch_size=16,
                                       class_mode='binary')
```

# Training Model With CNN

```python

cnn=tf.keras.models.Sequential()

cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64,64,3] ))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=2))

cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=2))

cnn.add(tf.keras.layers.Flatten())

cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
cnn.add(tf.keras.layers.Dense(units=64,activation='relu'))

cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

# Fitting model to Training Simultaneously Evaluating a Model

```python
cnn.fit(x=training_set, validation_data=test_set, epochs=50)
```

```python
training_set.class_indices
```

## Testing our model into validation images

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('../input/beauty-detection-data-set/valid/beautiful/008.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
      prediction = 'beutiful'
else:
      prediction = 'average'
    
print(prediction)
```
